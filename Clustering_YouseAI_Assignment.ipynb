## K MEANS ALGO

```python
from sklearn.cluster import KMeans
```

```python
# ELBOW METHOD (MANUAL METHOD) --> To select the k values

wcss=[]

for k in range(1,11):
    km=KMeans(n_clusters=k, init='k-means++')
    km.fit(pca_scaled)
    wcss.append(km.inertia_)
```

```python
# ELBOW CURVE

plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
km=KMeans(n_clusters=4, init='k-means++')
```

```python
# we use fit_predict because we have to find out labels for train_X data
km_labels=km.fit_predict(pca_scaled) # here we found out the labels of training data
```

```python
km_labels
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c= km_labels)
```

```python
# Performance Metric --> Silhouette score

from sklearn.metrics import silhouette_score
```

```python
kmeans_silhouette = silhouette_score(pca_scaled, km_labels)
```

```python
print("KMeans Silhouette Score:", kmeans_silhouette)
```

## Cluster Profiling

```python
# Cluster Profiling

df['KMeans_Cluster'] = km_labels
df['DBSCAN_Cluster'] = dbscan_labels
```

```python
# Grouping and describing clusters

kmeans_cluster_profile = df.groupby('KMeans_Cluster').mean()
dbscan_cluster_profile = df.groupby('DBSCAN_Cluster').mean()
```

#### Detailed Description

```python
# detailed descriptions of each segment

for cluster in kmeans_cluster_profile.index:
    print("Cluster",cluster ,":")
    print("Profile: \n", kmeans_cluster_profile.loc[cluster])
    print("Suggested Marketing Strategy: \n")
    if(cluster==3):
        break
    else:
        print("----------------------------- \n")
        
```

```python
for cluster in kmeans_cluster_profile.index:
    print("\nCluster", cluster,":")
    
    # Example of detailed descriptions (adjust based on your features)
    avg_balance = kmeans_cluster_profile.loc[cluster, 'BALANCE']
    avg_purchases = kmeans_cluster_profile.loc[cluster, 'PURCHASES']
    avg_credit_limit = kmeans_cluster_profile.loc[cluster, 'CREDIT_LIMIT']
    tenure = kmeans_cluster_profile.loc[cluster, 'TENURE']
    
    print("Average Balance:", round(avg_balance, 2))
    print("Average Purchases:", round(avg_purchases, 2))
    print("Average Credit Limit:", round(avg_credit_limit, 2))
    print("Tenure:", round(tenure, 2) , " months")
    
    # Interpret the cluster based on key metrics
    if avg_purchases > 1000:
        print("This segment consists of high-spending customers.")
    elif avg_purchases < 300:
        print("This segment consists of low-spending customers.")
    else:
        print("This segment consists of medium-spending customers.")
    
    if avg_credit_limit > 5000:
        print("These customers generally have a high credit limit.")
    else:
        print("These customers have a lower credit limit.")
    
    # Suggest a potential marketing strategy based on these insights
    if avg_balance > 1000:
        print("Marketing Strategy: Offer premium banking services.")
    else:
        print("Marketing Strategy: Promote basic banking offers.")
```

#### Marketing Strategies

1.For High Spenders:
Offer personalized rewards, exclusive offers, and access to premium services.

2.For Moderate Spenders:
Encourage higher usage with cashback or bonus points on specific categories (e.g., groceries, fuel)

3.For Low Spenders:
Introduce low-risk promotions like zero-interest offers for a limited time or educational campaigns.

Key Marketing Tactics by Segment:

1.Personalization: Tailor offers based on spending habits and preferences. Use personalized email campaigns, targeted advertisements, or app notifications to reach specific segments.

2.Rewards Programs: High and moderate spenders respond well to loyalty programs with cashback, discounts, and point accumulation.
Upselling: Encourage moderate spenders to upgrade their products (e.g., credit cards with better rewards, higher limits).

3.Educational Campaigns: For low spenders or infrequent users, educate them on the benefits of credit use or digital banking through blogs, emails, or customer service support.

4.Incentives: For segments with low usage or engagement, provide short-term rewards (e.g., cashbacks, discounts) to motivate higher engagement.
